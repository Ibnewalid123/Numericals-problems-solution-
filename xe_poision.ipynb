```python
import matplotlib.pyplot as plt
import numpy as np

fission_yeild_xe=0.0024
fission_yeild_I=0.0629
decay_const_xe=2.11*10**-5
decay_const_I=2.93*10**-5

fission_cross_section=543.2*(10**-24)*19.1 *6.023*10**23/235
absorbtion_cross_section_xe=1.61*10**5*(10**-24)

flux=5*10*14


def dXe():

    dxt=decay_const_I*I_o+fission_yeild_xe*flux*fission_cross_section-decay_const_xe*Xe_o-absorbtion_cross_section_xe*flux*Xe_o
    
    return dxt

def dI():

    dI=fission_yeild_I*fission_cross_section*flux-decay_const_I*I_o
    return dI

dt=10

T=24*3600*10
t=0

Xe=[]
I=[]
time=[]

I_o=0
Xe_o=0
while t<T:
    Xe_o=dXe()*dt+Xe_o
    I_o=dI()*dt+I_o
    time.append(t)
    Xe.append(Xe_o)
    I.append(I_o)
    t=t+dt

    if t>24*3600*2:
        flux=0

```

```python
time=np.array(time)/(24*3600)
plt.plot(time,Xe)
plt.plot(time,I)
plt.legend(['Xe','I'])
```
